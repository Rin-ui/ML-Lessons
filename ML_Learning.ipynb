```python
# label encoding --> gender
from sklearn.preprocessing import LabelEncoder
df_encoded = df.copy()
label_encoder = LabelEncoder()
df_encoded['gender'] = label_encoder.fit_transform(df_encoded['gender'])
df_encoded
```

```python
# label encoding --> relevent_experience
from sklearn.preprocessing import LabelEncoder
df_encoded = df.copy()
label_encoder = LabelEncoder()
df_encoded['relevent_experience'] = label_encoder.fit_transform(df_encoded['relevent_experience'])
df_encoded
```

```python
df_encoded['gender_encoded'] = label_encoder.fit_transform(df_encoded['gender'])
df_encoded['relevent_experience_encoded'] = label_encoder.fit_transform(df_encoded['relevent_experience'])
df_encoded['enrolled_university_encoded'] = label_encoder.fit_transform(df_encoded['enrolled_university'])
df_encoded
```

```python
# label encoding with order
education_level_encoded_dict = {"Primary School	":0, "High School":1, "Graduate":2, "Masters":3 }
df_encoded["education_level_encoded"] = df_encoded["education_level"].map(education_level_encoded_dict)
df_encoded
```

```python
# one hot encoding -> city
df_encoded = pd.get_dummies(df_encoded, columns=['city'], prefix='city')
df_encoded
```

```python
df_encoded = pd.get_dummies(df_encoded, columns=["major_discipline"], prefix="major_discipline")
df_encoded
```

```python
# onr hot encoding -> gender
df_encoded = pd.get_dummies(df_encoded, columns=["gender"], prefix="gender")
df_encoded
```

```python
# frequency encoding --> city
from sklearn.preprocessing import LabelEncoder
df_encoded = df.copy()
city_freq = df_encoded["city"].value_counts().to_dict()
df_encoded["city_count_encoded"] = df_encoded["city"].map(city_freq)
df_encoded
```

```python
unique_city = df_encoded["city"].unique()
unique_city
```

```python
# label encoding --> gender
from sklearn.preprocessing import LabelEncoder
df_encoded = df.copy()
label_encoder = LabelEncoder()
df_encoded['gender_encoded'] = label_encoder.fit_transform(df_encoded['gender'])
df_encoded
```

```python
# target encoding of education (feature) and target is grnder(it's avg we have to calculate)
gender_target_mean = df_encoded.groupby("education_level")["gender_encoded"].mean()
gender_target_mean
```

```python
# target encoding of education (feature) and target is grnder(it's avg we have to calculate)
gender_target_mean = df_encoded.groupby("education_level")["gender_encoded"].mean().to_dict()
df_encoded["gender_target_education_encoded"] = df_encoded["education_level"].map(gender_target_mean)
df_encoded
```

```python
# label encoding with order
education_level_encoded_dict = {"Primary School	":0, "High School":1, "Graduate":2, "Masters":3 }
df_encoded["education_level_encoded"] = df_encoded["education_level"].map(education_level_encoded_dict)
df_encoded
```

```python
# target education , feature gender
education_target_mean = df_encoded.groupby("gender")["education_level_encoded"].mean()
education_target_mean
```

```python
# target education , feature gender
education_target_mean = df_encoded.groupby("gender")["education_level_encoded"].mean()
df_encoded["education_target_gender_encoded"] = df_encoded["gender"].map(education_target_mean)
df_encoded
```
